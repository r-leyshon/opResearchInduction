```python
import os
from pyprojroot import here
import pandas as pd
```

```python
mara = pd.read_excel(
    os.path.join(
        here(), "extract", "Database Management and Descriptive Analytics", "Marathon.xlsx"
        )
        )
mara.head()
```

```python
sub_210 = mara[mara.Time >= 210]
grouped = sub_210.groupby("Gender")
sub_210_mean = grouped.agg("mean")
print(sub_210_mean)
sub_210_mean.plot.bar()
```

```python
sub_210_count = grouped.agg("count").drop(columns=["Place", "Age", "City", "Time"])
print(sub_210_count)
sub_210_count.plot.bar()
```
